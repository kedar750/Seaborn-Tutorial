# Catplot-Violinplot

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2.Import Data

```python
df = pd.read_csv('TRAIN (1).csv')
df.head()
```

```python
sns.catplot('Store_Type','Sales',data = df,hue = 'Region_Code',kind = 'violin')
```

```python
sns.catplot('Store_Type','Sales',data = df,hue = 'Region_Code',kind = 'violin',hue_order =  ['R4','R3','R1','R2'])
```

```python
sns.catplot('Store_Type','Sales',data = df,hue = 'Region_Code',kind = 'violin',inner =  'stick')
```

```python
sns.catplot('Store_Type','Sales',data = df,col = 'Region_Code',kind = 'violin')
```

```python
sns.catplot('Store_Type','Sales',data = df,col = 'Region_Code',kind = 'violin',col_order = ['R4','R3','R1','R2'])
```

# catplot-Barplot

```python
sns.catplot('Region_Code','Sales',data = df,kind = 'bar')
```

```python
sns.catplot('Region_Code','Sales',data = df,kind = 'bar',hue = 'Store_Type')
```

```python
sns.catplot('Region_Code','Sales',data = df,kind = 'bar',hue = 'Store_Type',col = 'Store_Type')
```

```python
sns.catplot('Region_Code','Sales',data = df,kind = 'bar',hue = 'Store_Type',col = 'Store_Type',col_wrap = 2,palette = 'rocket')
```

```python
sns.catplot('Region_Code','Sales',data = df,kind = 'bar',hue = 'Store_Type',row = 'Store_Type')
```
