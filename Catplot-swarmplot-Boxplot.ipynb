# Catplot-Swarmplot-Boxplot

# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2.Import Data

```python
df = pd.read_csv('TRAIN (1).csv')
df.head()
```

- Catplot is used to plot bar within how points were distributed.
- using catplot we can plot all differnt type of structures, that will see in the upcoming notebook.

```python
sns.catplot('Store_Type','Sales',data = df)
```

```python
sns.catplot('Store_Type','Sales',data = df,jitter = False)
```

```python
#sns.catplot('Store_Type','Sales',data = df,kind = 'swarm')
```

```python
sns.catplot('Store_Type','Sales',data = df,kind = 'box')
```

```python
sns.catplot('Store_Type','Sales',data = df,kind = 'box',hue = 'Location_Type')
```

```python
sns.catplot('Store_Type','Sales',data = df,kind = 'box',col = 'Location_Type')
```

```python
sns.catplot('Store_Type','Sales',data = df,kind = 'box',row = 'Location_Type')
```
